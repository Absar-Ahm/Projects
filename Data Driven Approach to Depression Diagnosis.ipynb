```python
# Importing the necessary libraries to perform our tasks
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Loading the dataset
from google.colab import drive
drive.mount('/content/drive/', force_remount=True)
df = pd.read_csv('/content/drive/MyDrive/ML/Minor Project - Form responses.csv')

# Preprocessing the data

# Identify non-numeric columns
non_numeric_columns = df.select_dtypes(include=['object']).columns

# Convert non-numeric columns using LabelEncoder
label_encoder = LabelEncoder()
for column in non_numeric_columns:
    df[column] = label_encoder.fit_transform(df[column])



# Removing rows with missing values
df = df.dropna()

# Defining the input features (X) and target variable (y) for our model
X = df.drop('Depression', axis=1)
y = df['Depression']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=98)

# Creating a logistic regression object
logreg1 = LogisticRegression()

# Training the model using the training data
logreg1.fit(X_train, y_train)

# Predicting the target variable for the testing data
y_pred = logreg1.predict(X_test)

# Evaluating the model's accuracy
accuracy = accuracy_score(y_test, y_pred)

# Evaluating the model's precision
precision = precision_score(y_test, y_pred)

# Evaluating the model's recall
recall = recall_score(y_test, y_pred)

# Evaluating the model's F1_score
f1 = f1_score(y_test, y_pred)

# Calculating Confusion Matrix
conf_mat = confusion_matrix(y_test, y_pred)

print("Accuracy: {:.2f}".format(accuracy))
print("Precision: {:.2f}".format(precision))
print("Recall: {:.2f}".format(recall))
print("F1 score: {:.2f}".format(f1))
print("Confusion matrix:\n", conf_mat)



```

```python
import seaborn as sns

# Creating heatmap of confusion matrix
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues')

# Set labels and title
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')

# Show plot
plt.show()
```
